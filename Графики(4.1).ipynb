```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import csv

x, y, districts = [], [], []

# Открытие файла с данными о ценах на недвижимость с сайта nb.ru и считывание информации
# Загрузка данных в переменные
with open("price_1.csv") as file:
    reader = csv.reader(file, delimiter=';')
    for line in reader:
        districts.append(line[0])
        x.append(int(line[1]))
        y.append(int(line[2]))


xy = range(len(districts))

# Помещаем кортеж, содержащий фигуру и объект осей в переменные fig и ax
fig, ax = plt.subplots()
# Отоборажаем отдельно точки однокомнатных и двухкомнатных квартир
plt.scatter(xy, x, label=u'One-room', color='b')
plt.scatter(xy, y, label=u'Two-room', color='r')
# Подписываем оси
plt.xlabel('District')
plt.ylabel('Cost')
# Устанавливаем метки для осей
plt.xticks(xy, districts)
# Форматируем
fig.autofmt_xdate(rotation=45)
plt.legend()

plt.show()
```
